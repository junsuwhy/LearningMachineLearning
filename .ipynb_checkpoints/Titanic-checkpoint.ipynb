# 讀 csv , GroupBy 用法

```python
df = pd.read_csv('titanic/train.csv');
df
```

```python
print(df.groupby(by='Sex').count()[:][['Survived']])
print(df.groupby(by='Sex').sum()[:][['Survived']])


group_by_sex_count = df.groupby(by='Sex').count()
group_by_sex_sum = df.groupby(by='Sex').sum()

# 注意資料型態，整數 / 整數還是整數，但我們要的是小數
# 要轉成字串，才能跟其他字串合併

print('女性生存機率：'+str(float(group_by_sex_sum.loc["female",'Survived']) / group_by_sex_count.loc["female",'Survived']))
print('男性生存機率：'+str(float(group_by_sex_sum.loc["male",'Survived']) / group_by_sex_count.loc["male",'Survived']))

group_by_pclass_count = df.groupby(by='Pclass').count()
group_by_pclass_sum = df.groupby(by='Pclass').sum()
group_by_pclass = df.groupby(by='Pclass')
print(group_by_pclass.size())

print('1 生存機率：'+str(float(group_by_pclass_sum.loc[1,'Survived']) / group_by_pclass_count.loc[1,'Survived']))
print('2 生存機率：'+str(float(group_by_pclass_sum.loc[2,'Survived']) / group_by_pclass_count.loc[2,'Survived']))
print('3 生存機率：'+str(float(group_by_pclass_sum.loc[3,'Survived']) / group_by_pclass_count.loc[3,'Survived']))
```

# 改變 columns 名稱、合併（concat, join）

```python
import matplotlib.pyplot as plt
gb_age_count = df.groupby(by='Age').count().loc[:,['Survived']]
gb_age_count.columns = ['count']
gb_age_sum = df.groupby(by='Age').sum().loc[:,['Survived']]
gb_age_sum.rename(columns={"Survived":"surv"}) #無效 ....
gb_age_sum

gb_age = pd.concat([gb_age_count,gb_age_sum], axis=1)
print(gb_age)
```

# 選特定columns 的資料，轉成 Array

```python
# 挑選資料
# df_simple = df.loc[:,['Pclass','Sex','Age','SibSp','Parch']]
df_simple = df.loc[:,['Pclass','Sex','Age','SibSp','Parch','Embarked']]
df_simple.loc[df_simple['Sex']=='male'] = 1
df_simple.loc[df_simple['Sex']=='female'] = 0
df_simple.loc[df_simple['Embarked']=='S'] = 0
df_simple.loc[df_simple['Embarked']=='C'] = 1
df_simple.loc[df_simple['Embarked']=='Q'] = 2
df_simple.values.tolist()
y = df.loc[:,['Survived']].values.tolist()
y
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets

# 建立模型
model = LinearRegression()
X = df_simple.values.tolist()
y = df.loc[:,['Survived']].values.tolist()
model.fit(X,y)
predict = model.predict(X)


predict_df = pd.DataFrame(predict, columns=["predict"])
predict_df.loc[predict_df['predict'] >= 0.5] = 1
predict_df.loc[predict_df['predict'] < 0.5] = 0

# 預測的跟結果的合在同一個 dataFrame
# predict_df = df.loc[:,['Survived']].append(pd.DataFrame(predict), ignore_index=True);
result_df = pd.concat([df.loc[:,['Survived']],predict_df], axis=1);

# 看有哪些猜錯
result_df.loc[result_df['Survived'] != result_df['predict']]
result_df.loc[result_df['Survived'] != result_df['predict']].count
# predict_df = pd.DataFrame({'survived':y,'predict':predict})
# predict_df = pd.DataFrame([np.array(y),predict], columns=['survived','predict'])
# result_df
```
