```python
from __future__ import division

def mean(X):
    return sum(X)/len(X)

ls = [1,2,3,4]
print mean(ls)
```

```python
def subtrack(X,Y):
    return [(y-x) for x,y in zip(X,Y)]

ls2 = [4,3,2,1]
print subtrack(ls, ls2)
```

```python
def dot(X,Y):
    return sum([(x * y) for x,y in zip(X,Y)])

print dot(ls, ls2)
```

```python
def de_mean(X):
    """把 X 轉成每個元素與平均值的差值"""
    x_bar = mean(X)
    return [x_i - x_bar for x_i in X]

print de_mean(ls)
```

```python
def sum_of_square(X):
    """每個元素的平方合"""
    return sum([x_i**2 for x_i in X])

print sum_of_square(ls)
```

```python
import math

def std(X):
    return math.sqrt(dot(X,X))

def correlation(X,Y):
    return dot(de_mean(X), de_mean(Y)) / (std(de_mean(X)) * std(de_mean(Y)))

print correlation(ls, ls2)
```

# 載入 titanic

```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic/train.csv');
df
```

# 看各變數的關聯

```python
"""看存活率跟艙級關聯"""
ls_sur = list(df['Survived'])
ls_pc = list(df['Pclass'])
print correlation(ls_sur, ls_pc)
```

```python
"""看存活率跟年紀關聯"""
"""比較麻煩是 age 有  NaN ，所以不能全部抓來看"""
df_age = df.loc[df['Age'] > 0].loc[:, ['Age', 'Survived']]

print correlation(list(df_age['Age']),list(df_age['Survived']))

# ls_age, = [(float(age),sur) for age, sur in df.loc[df['Age'] > 0]['Age', 'Survived']]

# print ls_age

# print correlation(ls_sur, ls_age)
```

```python
"""看性別的相關，可以看到女性比較容易存活"""
def is_male(sex):
    if(sex == 'male'):
        return 1
    else:
        return 0


ls_sex = [is_male(s) for s in df['Sex']]

print correlation(ls_sur, ls_sex)
```

```python
"""畫圖看看"""
import matplotlib.pyplot as plt 
plt.scatter(ls_sur, ls_sex)
plt.show()
```

```python
ls_fare = df['Fare']

print correlation(ls_pc, ls_fare)
```

```python
ls_fare_per_person = [(fare / (sibsp+1)) for fare, sibsp in zip(df['Fare'],df['SibSp'])]

print correlation(ls_pc, ls_fare_per_person)
```
